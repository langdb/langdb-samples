# Working with Postgres, MySQL, MongoDB, and Others
With LangDB, you can seamlessly connect to your preexisting databases, including Postgres, MySQL, MongoDB, and many others. This capability allows you to leverage your existing data infrastructure without the need for significant modifications.

## PostgreSQL

```sql
CREATE TABLE [IF NOT EXISTS] table_name
(
    col_name1 type1 ,
    col_name2 type2 ,
    ...
) ENGINE = PostgreSQL(host:port, database, table, user, password)
```

## MySQL

```sql
CREATE TABLE [IF NOT EXISTS] table_name 
(
    col_name1 type1 ,
    col_name2 type2 ,
    ...
) ENGINE = MySQL(host:port, database, table, user, password)
```

## MongoDB

```sql
CREATE TABLE [IF NOT EXISTS] table_name
(
    col_name1 [type1],
    col_name2 [type2],
    ...
) ENGINE = MongoDB(host:port, database, collection, user, password );
```

Once connected., you can power your databases using LangDB.
