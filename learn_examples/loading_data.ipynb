# Loading Data

Before your chosen AI provider can act on your data, you must load it. Proper data loading is crucial as it forms the backbone of any RAG  pipeline. Inferring the correct semantic meaning from unstructured data for downstream processing is a significant challenge. LangDB simplifies this by allowing you to load both unstructured and structured data from various sources, including files, databases, and images. This page will guide you through the process of loading data from these different sources.


## Documents (PDFs, HTMLs etc):

LangDB enables you to extract data from unstructured sources such as PDF documents and HTML files. Additionally, it offers built-in OCR support to extract text from images within PDF files. Prior to loading the data and generating embeddings, LangDB breaks down the documents semantically into chunks. This ensures that embeddings are generated for meaningful segments of the document. Alternatively, you can also specify your custom settings for chunking. 

The `text_split_options` parameter is an optional parameter in SQL Functions like [load_pdf_text()](../sql_functions/load_pdf_text.md) and [load_html()](../sql_functions/load_html.md) that allows you to specify options for splitting the files into chunks. The following options are available:

- `chunk_size` : The chunk capacity to use for filling chunks.

- `chunk_overlap`: The amount of overlap between chunks. Defaults to 0.

- `trim`: Specify whether chunks should have whitespace trimmed from the beginning and end or not. If false (default), joining all chunks should return the original string. If true, all chunks will have whitespace removed from beginning and end.


## Extracting PDF Text:

Ideally, users would like to start with creating a table to store the data extracted from the documents. The table can be created using the `CREATE TABLE` SQL command. 

```sql
CREATE TABLE pdfs (
  id UUID DEFAULT generateUUIDv4(),
  content `String`, 
  metadata `String`,
   
) 
engine = MergeTree
order by (id, content);
```

Once the table is created, the data can be loaded into the table using the `load_pdf_text` or `load_html` SQL functions. Files can be present locally/remote. 

```sql
INSERT INTO pdfs(content, metadata)
SELECT content, metadata from 
load_pdf_text ('https://d18rn0p25nwr6d.cloudfront.net/CIK-0000320193/faab4555-c69b-438a-aaf7-e09305f87ca3.pdf');
```

You can verify that data has been loaded into the table by running simple SQL.select count(*) from pdfs;

```sql
select count(*) from pdfs;
```

## Extracting HTML Text

Similarly, you can also extract text from HTML pages using the [load_html](../sql_functions/load_html.md) SQL function.

Example:

```sql
CREATE TABLE html_pages (
  id UUID DEFAULT generateUUIDv4(),
  content `String`, 
  metadata `String`,
   
) 
engine = MergeTree
order by (id, content);
```

```sql
INSERT INTO html_pages(content, metadata)
SELECT content, metadata from 
load_html('https://en.wikipedia.org/wiki/New_Delhi');
```
